### Реализация стохастического градиентного спуска

```python
def stochastic_gradient_descent(X, y, learning_rate=0.01, iterations=10000):
    X = X.reshape(-1, 1)
    X = np.concatenate((X, np.ones((len(X), 1))), axis=1)
    m, n = X.shape
    theta = np.zeros(n)
    
    for _ in range(iterations):
        for i in range(m):
            random_index = np.random.randint(m)
            xi = X[random_index]
            yi = y[random_index]
            gradient = 2 * xi.T.dot(xi.dot(theta) - yi)
            theta = theta - learning_rate * gradient
    
    return theta
```

```python
theta = stochastic_gradient_descent(X, y)
print(theta)
```

```python
y1 = theta[0] + theta[1]*X
```

```python
plt.scatter(X, y, color='grey')
z = np.polyfit(X, y, 1)
p = np.poly1d(z)
plt.plot(X, p(X), color='red')
o = np.polyfit(X, y1, 1)
v = np.poly1d(o)
plt.plot(X, v(X), color='blue')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Generated Data')
plt.show()
```
